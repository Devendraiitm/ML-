```python
# -*- coding: utf-8 -*-
"""
Created on Sat Jul 20 19:52:02 2019

@author: DEVENDRA
"""

#%%
# Devendra IIT Madras
# loading data 
import pandas as pd
file=pd.read_csv('F:\ZS\data.csv')
# deleting first column

file.area_of_shot=pd.Categorical(file.area_of_shot) 
file["a"]=file.area_of_shot.cat.codes
file.shot_basics=pd.Categorical(file.shot_basics) 
file["b"]=file.shot_basics.cat.codes
file.range_of_shot=pd.Categorical(file.range_of_shot) 
file["c"]=file.range_of_shot.cat.codes
file.game_season=pd.Categorical(file.game_season) 
file["d"]=file.game_season.cat.codes
file.type_of_shot=pd.Categorical(file.type_of_shot) 
file["e"]=file.type_of_shot.cat.codes
file.type_of_combined_shot=pd.Categorical(file.type_of_combined_shot)
file["f"]=file.type_of_combined_shot.cat.codes
file=file.rename(columns={"home/away":'kya'}) 
file.kya=pd.Categorical(file.kya) 
file["g"]=file.kya.cat.codes


#%%
file=file.drop(["area_of_shot","shot_basics","range_of_shot","game_season","type_of_shot","type_of_combined_shot","kya","shot_id_number","Unnamed: 0","team_name","date_of_game","lat/lng"],axis=1 )
#%%
idt=[]
for i in range(file.shape[0]):
    idt.append(i+1)
file["shot_id_number"]=idt
file_1=file[pd.notnull(file['is_goal'])]  # file 1 training dataset
file_2=file[pd.isnull(file['is_goal'])]   # Prediction dataset
#%%
# splitting 
file_1=file_1.fillna(0)
X=file_1.drop('is_goal',axis=1)
Y=file_1['is_goal']
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test= train_test_split(X,Y,test_size=0.2, random_state=0)

#from sklearn.ensemble import RandomForestClassifier
#rfc = RandomForestClassifier()
#rfc.fit(X_train,Y_train)
#rfc.predict(X_test)
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X_train,Y_train,sample_weight=None)
reg.predict(X_test)

#%%
X1=file_2.drop('is_goal',axis=1)
X1=X1.fillna(0)
Y1=reg.predict(X1)
#%%
sub=X1['shot_id_number']
sub=sub.to_frame()
sub['is_goal']=Y1
sub = sub.astype({"shot_id_number": int, "is_goal": float})
sub.to_csv('submission.csv',mode = 'w',index=False)
#%%
#print(rfc.feature_importances_)
p=pd.read_csv('F:\ZS\sample_submission.csv')
#p.info()
#ind1=[]
#for i in range(10):
#    if sub.iloc[i]["shot_id_number"]  in p["shot_id_number"]:
#        print(i)
#%%
bla=p.set_index("shot_id_number").join(sub.set_index("shot_id_number"),lsuffix='_1',rsuffix='_2')
bla=bla.drop(['is_goal_1'],axis=1)
bla=bla.rename(columns={'is_goal_2':'is_goal'})  
bla=bla.round(1)  
bla.to_csv('submission1.csv')
```
